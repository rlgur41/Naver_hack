```python
import turtle

x_line = turtle.Turtle()
y_line = turtle.Turtle()

color_list = ['red', 'blue', 'green', 'yellow']
    
class Graph:
    def __init__(self):  
        print("[OK] INIT Successfully")
        
            
    def draw_graph(self, data_list):
        x_line.penup()
        x_line.setposition(-250, -240)
        x_line.pendown()
        for idx, i in enumerate(data_list):
            x_line.forward(120)
            x_line.left(90)
            x_line.forward(10)
            x_line.backward(10)
            x_line.right(180)
            x_line.forward(10)
            
            x_line.penup()
            x_line.forward(30)
            x_line.right(90)
            x_line.forward(10)
            x_line.pendown()
            
            x_line.write(data_list[idx][0])
            
            x_line.penup()
            x_line.backward(10)
            x_line.left(90)
            x_line.backward(30)
            x_line.pendown()
            
            x_line.backward(10)
            x_line.left(90)
        
        y_line.penup()
        y_line.setposition(-250, -240)
        y_line.pendown()
        y_line.left(90)
        
        for i in range(10):
            y_line.forward(50)
            y_line.left(90)
            y_line.forward(5)
            
            y_line.penup()
            y_line.forward(30)
            y_line.pendown()
            y_line.write(i * 200 + 200)
            
            y_line.penup()
            y_line.backward(30)
            y_line.pendown()
            
            y_line.backward(5)
            y_line.right(180)
            y_line.forward(5)
            y_line.backward(5)
            y_line.left(90)
            
    def decode_data(self, file_name):
        clean_data = [[]]
        try:
            fd = open(file_name, 'r')
        except:
            print("[Fail] open error")
    
        data_stream = fd.read().split('\n')
        for idx, i in enumerate(data_stream):
            clean_data.insert(idx, i.split('\t'))
            print(idx)
            
        clean_data.remove([])
        
        return clean_data
    
    def draw_chart(self, data_list):
        
        for idx, i in enumerate(data_list):
            t = turtle.Turtle()
            t.shape('turtle')
            t.fillcolor(color_list[idx])
            t.pencolor(color_list[idx])
            t.penup()
            t.setposition(-250,-240)
            t.forward(120 * idx + 120)
            t.pendown()
            t.left(90)
            
            move = int(i[1])
            
            t.forward(move//4)
            
            t_char = turtle.Turtle()
            t_char.penup()
            t_char.setposition(t.position())
            t_char.right(90)
            t_char.forward(30)
            t_char.left(90)
            t_char.forward(30)
            t_char.pendown()
            t_char.write(i[1])
             
            
g = Graph()

try:
    data_list = g.decode_data("C:/Users/rlgur/Documents/data_handled.txt")
except:
    print("Open error")

g.draw_graph(data_list)
g.draw_chart(data_list)

        
```
