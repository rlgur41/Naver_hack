```python
import re
import requests
import time


list_naver_search = []
list_check_data = []
print('start')  

while(True):
    data_file = open("D:\HardDisk\hadoop_data\data", 'a') 
    list_naver_search = re.findall('<span class="ah_k">(.*)</span>', requests.get('http://naver.com').text)[:20]
    
    for now_data in list_naver_search:
        if(now_data not in list_check_data):
            print(list_naver_search)
            data = (",".join(list_naver_search))
            
            list_check_data = list_naver_search
            
            data_file.write(data)
            break
            
    time.sleep(3)
    data_file.close()
    
```
